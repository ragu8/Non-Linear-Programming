```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## f(x) = 4 * x - 1.8 * x ** 2 + 1.2 * x ** 3 - 0.3 * x ** 4 

```python
def cal_f(x): 
    return 4 * x - 1.8 * x ** 2 + 1.2 * x ** 3 - 0.3 * x ** 4

def golden_section_search(func, xl, xu, es=1e-5, max_iter=100):
    R = (math.sqrt(5) - 1) / 2
    d = R * (xu - xl)
    x1 = xl + d
    x2 = xu - d
    f1 = func(x1)
    f2 = func(x2)
    iter_count = 0
    
    plt.figure(figsize=(10, 6))
    plt.plot([xl, xu], [func(xl), func(xu)], marker='o', label='Initial Interval')
    
    while abs(xu - xl) > es and iter_count < max_iter:
        if f1 > f2:
            xl = x2
            x2 = x1
            x1 = xl + (1 - R) * (xu - xl)
            f2 = f1
            f1 = func(x1)
        else:
            xu = x1
            x1 = x2
            x2 = xu - (1 - R) * (xu - xl)
            f1 = f2
            f2 = func(x2)
        
        plt.plot([xl, xu], [func(xl), func(xu)], marker='o', label='Updated Interval')
        
        iter_count += 1

    if f1 > f2:
        max_x, max_f = x1, f1
    else:
        max_x, max_f = x2, f2
    
    plt.scatter(max_x, max_f, color='red', label='Minimum Found')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Function and Minimum Found by Golden Section Search')
    plt.legend()
    plt.grid(True)
    plt.show()

    return max_x, max_f


def my_function(x):
    return 4 * x - 1.8 * x ** 2 + 1.2 * x ** 3 - 0.3 * x ** 4  

xl = 0  
xu = 4  
es = 1e-5  
max_iter = 100  

# Perform Golden Section Search
result, f_result = golden_section_search(my_function, xl, xu, es, max_iter)

print("Minimum found at x = {:.5f}, f(x) = {:.5f}".format(result, f_result))
```
