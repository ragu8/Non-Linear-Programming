```python
from scipy.integrate import odeint
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
# Differential equation: dy / dx = x, y(0) = 1, xrange = [0, 1]
def f1(y, x):
    return x

y0_1 = 1
xs_1 = np.arange(0, 1, 0.1)
ys_1 = odeint(f1, y0_1, xs_1)

# Plotting the first differential equation
plt.plot(xs_1, ys_1, '-')
plt.plot(xs_1, ys_1, 'ro')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Differential equation: dy / dx = x, y(0) = 1, xrange= [0,1]')
plt.show()
```

```python
# Differential equation: dy / dx = (x*y/(np.sqrt(y**2 + x))), y(0) = 1, xrange = [0, 5]
def f2(y, x):
    dydx = x * y / np.sqrt(y**2 + x)
    return dydx

y0_2 = 1
xs_2 = np.arange(0, 5, 0.25)
ys_2 = odeint(f2, y0_2, xs_2)

# Plotting the second differential equation
plt.plot(xs_2, ys_2, '-')
plt.plot(xs_2, ys_2, 'ro')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Differential equation: dy / dx = (x*y/(np.sqrt(y**2 + x))), y(0) = 1, xrange= [0,5]')
plt.show()
```

```python
# Differential equation: dy / dx = (1 - 2t)u, u(0) = 1, xrange = [-1, 2]
def f3(u, t):
    dudt = (1 - 2 * t) * u
    return dudt

u0_3 = 1
xs_3 = np.arange(-1, 2, 0.2)
ys_3 = odeint(f3, u0_3, xs_3)

# Plotting the third differential equation
plt.plot(xs_3, ys_3, '-')
plt.plot(xs_3, ys_3, 'ro')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Differential equation: dy / dx = (1 - 2t)u, u(0) = 1, xrange= [-1, 2]')
plt.show()
```
