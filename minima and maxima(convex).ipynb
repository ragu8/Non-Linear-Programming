```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import minimize_scalar
```

```python
f1 = lambda x : (x - 1) ** 4 + x ** 2
res = minimize_scalar(f1, method='brent')

print('xmin: %.02f, fval: %.02f, iter: %d' % (res.x, res.fun, res.nit))
x = np.linspace(res.x - 0.5, res.x + 0.5, 100)

y = [f1(val) for val in x]

plt.plot(x, y, color='blue', label='f1')
plt.title("(x - 1) ** 4 + x ** 2")
plt.scatter(res.x, res.fun, color='red', marker='x', label='opt')
plt.grid()
plt.legend()
plt.show()
```

```python
f2 = lambda x :  np.exp((x - 0.7)**2)
res = minimize_scalar(f2, method='brent')

print('xmin: %.02f, fval: %.02f, iter: %d' % (res.x, res.fun, res.nit))
x = np.linspace(res.x - 0.5, res.x + 0.5, 100)
y = [f2(val) for val in x]
plt.plot(x, y, color='blue', label='f1')
plt.title("np.exp((x - 0.7)**2)")
plt.scatter(res.x, res.fun, color='red', marker='x', label='opt')
plt.grid()
plt.legend()
plt.show()
```
